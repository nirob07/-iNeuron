```python
from sklearn.cluster import KMeans
np.random.seed(0)
X = face.reshape((-1, 1))
n_clusters = 5
k_means = KMeans(n_clusters=n_clusters, n_init=4) 
k_means.fit(X) 
values = k_means.cluster_centers_.squeeze() 
labels = k_means.labels_
len(labels)
```
