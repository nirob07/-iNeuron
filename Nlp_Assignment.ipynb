```python
import numpy as np
import nltk
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.php.net/"
res = requests.get(url)
if res.status_code == 200:
    html_doc = res.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
all_text = soup.get_text()
sentences = [k.strip() for k in all_text.split("\n") if k]
sentences = [l.strip() for l in sentences if l]
sentences
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokenized_sents = [word_tokenize(i) for i in sentences]
# for i in tokenized_sents:
#     print(i)
tokenized_sents
```

```python
import itertools
# final_list=list(itertools.chain(tokenized_sents))
# print('Merged List:')
# print(final_list)
flatten_list = list(itertools.chain.from_iterable(tokenized_sents))
flatten_list
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english", ignore_stopwords=True)
```

```python
flatten_list = [stemmer.stem(fl) for fl in flatten_list]
flatten_list
```

```python
stopwords=nltk.corpus.stopwords.words('english')
words_ne=[]
for word in flatten_list:
    if word.isalnum() and  word not in stopwords:
        words_ne.append(word)
```

```python
words_ne
```

```python
from nltk.probability import FreqDist
```

```python
data_analysis = nltk.FreqDist(words_ne)
data_analysis.plot(25, cumulative=False)
```

```python
data_analysis.most_common(15)
```

```python
data_analysis.tabulate(15)
```
