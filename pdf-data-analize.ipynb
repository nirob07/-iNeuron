```python
import tabula
csv_data = tabula.read_pdf("convertcsv.pdf", pages="all")
```

```python
df = pd.concat([csv_data[i] for i in range(len(csv_data))])
df["in"] = [i for i in range(len(df))]
df.set_index("in", inplace=True)
```

```python
df.to_csv("pdf_to_sv.csv",index=False)
```
